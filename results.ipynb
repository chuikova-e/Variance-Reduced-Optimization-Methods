```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
losses_CGS = np.load('losses/losses_CGS.npy')
losses_SFF = np.load('losses/losses_SFF.npy')
losses_SGD = np.load('losses/losses_SGD.npy')
losses_STORC = np.load('losses/losses_STORC.npy')
losses_SVRF = np.load('losses/losses_SVRF.npy')
times_CGS = np.load('losses/times_CGS.npy')
times_SFF = np.load('losses/times_SFF.npy')
times_SGD = np.load('losses/times_SGD.npy')
times_STORC = np.load('losses/times_STORC.npy')
times_SVRF = np.load('losses/times_SVRF.npy')
losses_reSGD = np.load('losses/losses_reSGD.npy')
times_reSGD = np.load('losses/times_reSGD.npy')
```

```python
plt.figure(figsize=(10,6))
plt.plot(times_CGS[:-1], losses_CGS, label = 'SCGS', color='aqua')
plt.plot(times_SFF[:50], losses_SFF[:50], label = 'SFW', color='b')
plt.plot(times_SGD[:30], losses_SGD[:30], label = 'SGD', color = 'black')
plt.plot(times_STORC[:20], losses_STORC[:20], label = 'STORC', color='lime')
plt.plot(times_SVRF, losses_SVRF, label = 'SVRF', color='red')
plt.plot(times_reSGD[:100], losses_reSGD[:100], label = 'SVRG', color = 'pink')
plt.legend()
plt.xlabel('time(sec)')
plt.ylabel('loss')
plt.savefig('result.jpg')
plt.show()


#plt.plot(times_CGS, losses_CGS, label = 'CGS')
```

```python
losses_FW = np.load('pagerank/losses_FW.npy')
times_FW = np.load('pagerank/times_FW.npy')
losses_GD = np.load('pagerank/losses_GD.npy')
times_GD = np.load('pagerank/times_GD.npy')
losses_GD_p = np.load('pagerank/losses_GD_polyak.npy')
times_GD_p = np.load('pagerank/times_GD_polyak.npy')
losses_SVRF = np.load('pagerank/losses_SVRF.npy')
times_SVRF = np.load('pagerank/times_SVRF.npy')
```

```python
plt.figure(figsize=(10,6))
plt.plot(times_FW, losses_FW, label = 'FW')
plt.plot(times_GD, losses_GD, label = 'GD')
plt.plot(times_GD_p, losses_GD_p, label = 'GD with Polyak step')
plt.plot(times_SVRF[:100], losses_SVRF[:100], label = 'SVRF')
plt.legend()
plt.xlabel('time(sec)')
plt.ylabel('loss')
plt.savefig('result_PR.jpg')
plt.show()


#plt.plot(times_CGS, losses_CGS, label = 'CGS')
```
